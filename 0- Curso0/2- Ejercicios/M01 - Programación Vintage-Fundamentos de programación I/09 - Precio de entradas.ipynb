# Determinar el precio de un conjunto de entradas al Zoo

Hacer un programa que muestre el precio total de las entradas de un grupo de personas al zoo. Teniendo en cuenta los siguientes precios:

- niños de 2 o menos años: no pagan
- Hasta los 12 años: 14 €
- Jubilados de 65 o más año: 18€
- Resto: 23€.

El programa pedirá la edad de los componentes del grupo, primero pedirá una edad, y sucesivamente las demás. En el momento en que se introduzca un 0 se considerará el final y el programa devolverá lo siguiente:

```
1 entradas de baby (0€):      0.00€
3 entradas de menores (14€): 42.00€
2 entradas normales (23€):   46.00€
1 entradas jubilado (18€):   18.00€
                             ------
                            106.00 €
```

````
#inicializar 

PRECIOS = (0, 14, 23, 18)
EDADES = (2, 12, 64, float("inf"))
CATEGORIA = ["baby", "niño", "adulto", "jubilado"] 

participante = 0
edadesGrupo = []
````

````

#pedir datos

print("Voy a calcular el precio de las entradas. Te pediré las edades de los participantes de uno en uno. Pulsa enter cuando hayas terminado.")

while True:
  edad = input("Dime la edad del participante {}: ".format(participante + 1))
  if edad == "":
    break
  try: 
    edad = int(edad)
    if edad < 0:
      print("La edad debe ser mayor que 0.")
    else:
      edadesGrupo.append(edad)
      participante += 1
  except:
    print ("La edad debe ser un número.") 

print("Ok, estas son las edades de los", participante, "participates: ", edadesGrupo)
````

````
# Calcular tabla, funciones:

def rango(edad): 
    for i in range(len(EDADES)): 
      if edad <= EDADES[i]:
        return i

def agrupar(edades):
  grupo=[0]*len(EDADES)

  for edad in edades: 
    grupo[rango(edad)] += 1  
  
  return grupo

def printfactura(edades):
  grupo = agrupar(edades)
  total = 0
  
  for i in range(len(EDADES)):
    if grupo[i] != 0:
      subtotal = grupo[i]*PRECIOS[i]
      total += subtotal 
      print("{} entradas de {}, {} €: \t {}".format(grupo[i], CATEGORIA[i], PRECIOS[i], subtotal))

  print ("\nTOTAL: {} €".format(total) )
````

````
# Calcular tabla:

printfactura(edadesGrupo)
````
