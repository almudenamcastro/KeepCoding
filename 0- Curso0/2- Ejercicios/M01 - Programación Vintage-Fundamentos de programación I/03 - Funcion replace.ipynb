# Crear funcion replace

Crear una función que tome una cadena y remplace el caracter de una posición dada por otro según la definición 

```
 myReplace(cadena, posicion, nuevoValor)
```

##Restricciones

- Teniendo en cuenta que en python una cadena es realmente una tupla de caracteres (inmutable, no se puede modificar ni su longitud, ni su contenido) deberás valorar crear una nueva cadena con el resultado y devolverla.
- Crear un módulo reutilizable. Para hacerlo correctamente (incluyendo la distinción entre utilizarlo como módulo o como main - no visto aún en clase - mira [aquí](https://drive.google.com/open?id=1bbnwN0ADgMtjNei0EI94NqXcCA1rKz9a))



````
def myReplace (cadena, posicion, nuevoValor):
  if posicion > len(cadena):
    raise ValueError("la posición debe ser menor que la longitud de la palabra") 
  cadena = list(cadena)
  cadena[posicion -1]=nuevoValor

  return "".join(cadena)

def myReplace2 (cadena, posicion, nuevoValor):
  if posicion > len(cadena):
    raise ValueError("la posición debe ser menor que la longitud de la palabra") 
  return cadena[:posicion-1]+nuevoValor+cadena[posicion:]
````

````
myReplace("paraguas", 3, "n")
````

## Retos.
- Crear la función dada como solución con un for
- Crear una función que no indique la posición sino con esta firma `myReplace(cadena, viejoCaracter, nuevoCaracter)` y que lo haga para todas las ocurrencias de `viejoCaracter`


````
def myReplace3(cadena, viejoc, nuevoc):
  resultado=""
  for letra in cadena:
    if letra == viejoc:
      resultado += nuevoc
    else:
      resultado += cadena[cadena.index(letra)]
  return resultado
````
