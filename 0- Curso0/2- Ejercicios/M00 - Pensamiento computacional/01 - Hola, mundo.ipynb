```
name = input("Hola, ¿cómo te llamas? ")
if name == "Lucas":
  print("hasta luego, %s" % name)
elif name[-3:] == "elo":
  print("hola %s, tu nombre rima con caramelo" % name)
else:
  print("encantade de saludarte, %s" % name)
```

```
import random

name = input("Hola, ¿cómo te llamas? ")

saludos = ["Hola,", "Un placer saludarte,", "Mucho gusto,", "Encantade,", "Buenas,", "Qué bonito nombre,"]
print(random.choice(saludos), name)

print(saludos[random.randint(0, len(saludos))], name)
```
