```python
import pandas as pd
```

Cargar datos

```python
''' 
0. Importamos el dataset de los datos de los pokemon
'''
df = pd.read_csv("pokemon.csv")
df.head()
df.tail()
```

```python
''' 
1. ¿Cuántos pokémon hay en el dataset?
''' 

df['Name'].count()
len(df)
```

```python
'''
2. ¿Cuántos pokémon hay de tipo Poison? (Type 1)
'''
# filtro el data frame y le pregunto cuántos nombres hay.  
df[df['Type 1']=='Poison']['Name'].count()
```

```python
'''
3. ¿Cuántos tipos diferentes de pokémon hay? (Type 1)
'''
len(df.groupby('Type 1'))

# también se puede calcular con unique count: 
df['Type 1'].nunique()
```

```python
'''
3.1. ¿Cuántos pokémon hay de cada tipo? (Type 1)
'''

df.groupby('Type 1')['Name'].count()
```

```python
'''
3.2 ¿cual es el tipo de pokémon con más pokémon? (Type 1)
'''

# agrupamos por tipo, contamos los nombres, ordenamos descendentemente y nos quedamos con el primer índice (nombre)
df.groupby('Type 1')['Name'].count().sort_values(ascending = False).index[0]

#también podemos guardar la serie resultante de la agregación y filtrarla por el valor máximo: 
tipos = df.groupby('Type 1')['Name'].count()
tipos[tipos == tipos.max()]
```

```python
'''
3.3. ¿Cuál es el pokémon más rápido?
'''

df[df['Speed']==df['Speed'].max()]
```

```python
'''
4. ¿Cuántos pokemon tiene una defensa superior a 100?
'''

len(df[df['Defense']>100])
```

```python
'''
5. ¿Cuántos pokemon tiene una defensa superior a 100 y una velocidad superior a 100?
'''
len(df[(df['Defense']>100) & (df['Speed']>100)])
```

```python
'''
6. Ordena el dataset por el tipo 1 y por el tipo 2
'''
df.sort_values(['Type 1', 'Type 2'])
```

```python
'''
7. Crea un nuevo dataset con los pokémon de tipo Water y Fire como primer tipo
'''

fire_water = df[(df['Type 1'] == 'Water') | (df['Type 1'] == 'Fire') ]
fire_water
```

```python
''' 
8. Crea un nuevo dataset con los pokemon Legendary
'''

legendary = df[df['Legendary']]
legendary
```

```python
'''
9. Calcula el promedio de stats de los pokemon Legendary (HP, Attack, Defense, Sp. Atk, Sp. Def, Speed) y los no Legendary
'''
# los legendary (el resultado es una serie)
df[df['Legendary']][['HP','Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].mean()
# los no legendary
df[~df['Legendary']][['HP','Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].mean()

# también se podrían calcular con describe: 
df[df['Legendary']].describe().loc['mean', ['HP','Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
df[~df['Legendary']].describe().loc['mean', ['HP','Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
```

```python
'''
10. Crea un nuevo dataframe con el resultado del anterior ejercicio comparando ambos promedios

 ejemplo:
                HP  Attack  Defense  Sp. Atk  Sp. Def  Speed
 Legendary       99      90       89       91       94     90
 No Legendary    80      95       75       12       43     87
'''
# junto las dos series en un df, indicando los índices "Legendary" y "No Legendary"
df2 = pd.DataFrame([df[df['Legendary']].describe().loc['mean'][2:-1], df[~df['Legendary']].describe().loc['mean'][2:-1]], index=['Legendary', 'No Legendary'])
df2
```

```python
from functools import reduce

# 13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
# Puedes definir más funciones si lo consideras necesario.

articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']
```

```python
# A. Una función que calcule el precio total de la compra

precio = lambda iter: reduce (lambda x,y:x+y, map(lambda x:articulos[x], iter))
precio (compra)
```

```python
# B. Una función que calcule el precio total de la compra con un descuento del 10%
precio_d10 = lambda iter: precio(iter)*0.9
precio_d10(compra)
```

```python
# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
precio_desc_cond = lambda iter: precio_d10(iter) if precio(iter) >100 else precio(iter)
precio_desc_cond(['Gafas', 'Gafas', 'Gafas', 'Gafas', 'Gafas'])
```

```python
# D. Una función que calcule el precio total aplicando el IVA (21%)

precio_IVA = lambda iter: precio(iter)*1.21
precio_IVA(compra)
```

```python
# E. lista los artículos cuyo precio es superior a 20€
caros = lambda iter: filter(lambda x: articulos[x] > 20, iter)
list(caros(compra))
```

```python
'''
14. Dada una lista de tuplas con el nombre de un alumno, apellidos, curso y sus notas. 

 Define una funcion que reciba el curso y saque una lista en la que aparezca e nombre y apellidos y el promedio de sus notas.
 Puedes definir más funciones si lo consideras necesario.
'''

alumnos = [('Juan', 'Pérez', '1', [5, 6, 7, 8, 9]),
            ('Ana', 'García', '2', [5, 6, 7, 8, 9]),
            ('Luis', 'González', '1', [5, 6, 7, 8, 9]),
            ('María', 'Martínez', '2', [5, 6, 7, 8, 9]),
            ('Pedro', 'Rodríguez', '1', [5, 6, 7, 8, 9]),
            ('Lucía', 'Hernández', '2', [5, 6, 7, 8, 9]),
            ('Marta', 'López', '1', [5, 6, 7, 8, 9]),
            ('Sara', 'Díaz', '2', [5, 6, 7, 8, 9]),
            ('Javier', 'Sánchez', '1', [5, 6, 7, 8, 9]),
            ('Miguel', 'Fernández', '2', [5, 6, 7, 8, 9]),
            ('Sergio', 'Jiménez', '1', [5, 6, 7, 8, 9]),
            ('Sandra', 'Ruiz', '2', [5, 6, 7, 8, 9]),
            ('Pablo', 'Álvarez', '1', [5, 6, 7, 8, 9]),
            ('María', 'Gómez', '2', [5, 6, 7, 8, 9]),
]

# Usando filter y map:

filtrar_por_curso = lambda curso, lista: filter(lambda x: x[2] == curso, lista)
calcular_media = lambda lista: map(lambda x: (x[0]+" "+x[1], sum(x[3])/len(x[3])), lista)

def media_filtrada_v1(curso, lista = alumnos):
    return list(calcular_media(filtrar_por_curso(curso, lista)))


# También se podría resolver usando list comprehensions: 

def media_filtrada_v2 (curso, lista=alumnos): 
    return [(x[0]+" "+x[1], sum(x[3])/len(x[3])) for x in lista if x[2] == curso ]

# Compruebo que el output de las dos funciones es el mismo: 

print("\ncon funciones: \ncurso 1:\n", media_filtrada_v1("1"), "\ncurso 2:\n", media_filtrada_v1("2"))

print("\ncon list compregensions: \ncurso 1:\n", media_filtrada_v2("1"), "\ncurso 2:\n", media_filtrada_v2("2"))
```
