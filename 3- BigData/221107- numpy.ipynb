### Primeros pasos

- a = np.array() -> para crear un array. 
- a.ndim -> para saber cuántas dimensiones tiene. 
- a.shape -> para saber la longitud de cada dimensión. **texto en negrita**

```python
''' 
llamamos a np.array() para iniciar un array en numpy
'''

a = np.array([1,2,3,4,5])
print("de una dimensión:\n", a, "\n")

# de dos
b =[[1,2,3],[4,5,6]]
print("de dos dimensiones:\n", b, "\n")

# o más
c = np.array([
    [
        [1,2,3],[4,5,6]
    ],
    [
        [1,2,3],[4,5,6]
    ]   
    ])
print("o más:\n", c, "\n")
```

```python
'''
podemos usar dtype para ajustar el tipo de datos almacenandos y optimizar memoria. 
'''
print("por defecto, una matriz guarda datos de tipo", a.dtype)
a32 = np.array(a*10000, dtype = np.int32)
print("podemos definir a32:", a32, "con datos de tipo", a32.dtype, "\n")

'''
podemos redefinir el tipo de dato de una matriz con astype. 
astype no modifica la función original (hay que guardarlo en otra variable). 
Pero ojo, si metemos algo que se salga en memoria va a dar error (ej: un número muy grande).
Ej: como máximo int16 
'''
#a16 = a32.astype(np.int16)
a16 = np.array([1,2,32770], dtype=np.int16) 

print(a16, "parece que se ha roto")
```

```python
'''
Para obtener la información de una matriz, usamos: 
- ndim para obtener las dimensiones de un array
- shape para obtener su tamaño (shape devuelve una tupla).
- size para saber cuántos datos almacena. 
- itemsize para saber cuánto ocupa en memoria cada dato. 

'''
print("la matriz tiene", c.ndim, "dimensiones")
print("cada dimensión mide", c.shape)
print("guarda", c.size, "datos")
print("ocupa", c.itemsize, "bytes por dato")
print("en total: ", c.itemsize*c.size, "bytes")
```

```python
print("elementos de a16:", a16.size)
print("tamaño de cada elemento:", a16.itemsize)
print ("tamaño total:", a16.itemsize * a16.size)
```

#### Matriz de ceros

```python
# Definir una matriz 
# primer arg -> tamaño (se introduce el shape con una tupla) con ceros. 
# argumento optional -> se puede indicar el tipo de dato con dtype
# np.zeros_like(matriz) -> toma la forma y el tipo de dato de otra matriz
print("matriz np.zeros(shape, dtype)\n", np.zeros((2,3,5), dtype=np.int32), "\n")
print("matriz np.zeros(b.shape)\n", np.zeros(b.shape), "\n")
print("matriz np.zeros_like(b)\n", np.zeros_like(b), "\n")
```

Matriz de valores concretos

```python
'''
usamos np.full(shape, value). 
También acepta como argumento dtype
Podemos usar un valor acabado en . para que np sepa que queremos un float. 
'''
p = np.full((5,3), 50.)
print(p)
```

### Álgebra lineal

numpy py permite realizar operaciones matriciales como por elemplo:
- determinante:   np.linalg.det()
- inversa:        np.linalg.inv()
- producto        matricial: np.dot() o np.matmul() o a@b
- diagonal:       np.diag()
- inversa de una matriz: np.linalg.inv()



```python
'''
se puede definir una matriz identidad
'''
identidad = np.eye(5)
print(np.linalg.det(identidad))
```

```python
#para escribir una multiplicación matricial, podemos usar np.dot(a,b) o el signo arroba, que invoca a matmul: a@b
identidad @ identidad
```

Más tipos de operaciones con arrays:
- **np.sum():** Suma de todos los elementos de un array
- **np.mean():** Media de todos los elementos de un array
- **np.std():** Desviación estándar de todos los elementos de un array
- **np.var():** Varianza de todos los elementos de un array
- **np.min():** Valor mínimo de todos los elementos de un array
- **np.max():** Valor máximo de todos los elementos de un array
- **np.argmin():** Indice del valor mínimo de un array
- **np.argmax():** Indice del valor máximo de un array
- **np.cumsum():** Sumatoria acumulada de un array
- **np.cumprod():** Producto acumulado de un array
- **np.flip():** Inversa de un array
