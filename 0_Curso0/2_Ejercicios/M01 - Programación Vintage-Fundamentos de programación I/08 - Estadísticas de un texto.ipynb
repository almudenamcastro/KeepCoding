# Modulo de estadísticas de un texto

Crear un módulo con las siguientes funciones

- contarCaracteres(texto)
- contarPalabras(texto)
- contarVocales(texto)
- contarConsonantes(texto)

## Restricciones
- Cada función devolverá el valor numérico de lo que cuenta.
- Si el módulo se ejecuta directamente (sin ser importado por otro programa deberá listar todos los valores de un texto en forma de tabla tabulada)

## resolucón:

1.   contarCaracteres



```
def contarCaracteres(string):
  return len(string)
```

```
contarCaracteres("el perro del hortelano")
```

2.   contarPalabras



```
def contarPalabras1(string):
  if string =="" or string.isspace():
    return 0
  else: 
    palabras = 1
    for i in range(len(string)):
      if string[i]==" " and string[i+1].isalpha(): 
        palabras += 1
    return palabras

  def contarPalabras(string):
    return len(string.split(" "))
```

```
contarPalabras("el perro de don Ramón no tiene rabo")
```

3.   contarVocales

```
from unidecode import unidecode 

def contarVocales(string):
    string = unidecode(string)
    count = 0
    for letra in string: 
      if letra.lower() in "aeiou":
        count += 1
    return count
```

```
contarVocales("el murciélago no vuela")
```

3.   contarConsonantes




```
def contarConsonantes(string): 
  string = unidecode(string)
  count = 0
  for letra in string:
    if letra.isalpha() and letra not in "aeiou":
      count += 1
  return count
```

```
contarConsonantes("¡Manda narices!")
```

```
string = input("Dime una frase: ")
for funcion in [contarCaracteres, contarPalabras, contarVocales, contarConsonantes]: 
  print ("{}: \t {}".format(funcion.__name__.replace("contar", ""), funcion(string)))
```

## Retos

- Crear otros contadores como
  - contar signos de puntuacion fijando los mismos por ejemplo en los siguientes `.,;:¡!¿?()[]-_`
  - contar un carácter determinado con la función contar(texto, 'r')

```
def contarSignos(string):
  count=0
  for letra in string:
    if not(letra.isalnum() or letra.isspace()):
      count +=1
  return  count
```

```
contarSignos("¡¡el 3 es un ñnúmero bonito!! ¿? :.,")
```

```
def contarCaracter(string, caracter):
  count = 0
  string = unidecode(string).lower()
  caracter = caracter.lower()
  for i in string:
    if i == caracter:
      count += 1
  return count
```

```
contarCaracter("El perro de san Roque", "r")
```
