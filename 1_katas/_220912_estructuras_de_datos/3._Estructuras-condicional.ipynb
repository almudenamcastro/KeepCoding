## if ... else

```
if condicion:
    bloque si se cumple la condicion
else:
    bloque si NO se cumple la condición
```

````python
numero = 16

if numero % 2 == 0:
    print("El número es par")

else:
    print("El número es impar")
````

    bloque si se cumple la condición 3
...
else:
    bloque si NO se cumple ninguna condición

## Ejercicio 1

Escribir un programa que indique la generación correspondiente para un año de nacimiento indicado

<img src = "https://integrandoequipos.com/wp-content/uploads/2019/06/generaciones2.jpg">

````python
AGES = (1946, 1965, 1980, 2001, 2010)
GENERATION = ("Silenciosa", "baby boomer", "Gen. X", "Gen. Y", "Gen. Z")

def generation(year):
  if year < 1920: 
    return "Dime un año posterior a 1920"
  i=0
  for i in range(len(AGES)):
    if year < AGES[i]:
      return GENERATION[i]
    else: 
      i += 1
  return "No hay generaciones definidas para los nacidos después de 2010"
````

````python
generation(1987)
````
