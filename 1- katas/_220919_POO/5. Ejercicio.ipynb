## Consigna:

Realizar un programa que conste de una clase llamada Alumno que tenga como atributos el nombre y la nota del alumno. 
Definir los mÃ©todos para inicializar sus atributos, imprimirlos y mostrar un mensaje con el resultado de la nota y si ha aprobado o no.

```python
class Alumno:

  def __init__(self, name, grade):
    self.name = name
    self.grade = grade

  def __str__(self):
    return f"\nName: {self.name} \nNota: {self.grade}"

  def __repr__(self):
    return self.__str__() 

  def aprobado(self):
    if self.grade > 5:
      return True
    else:
      return False

```

```python
marta = Alumno ("Marta", 6)
```

```python
print(marta)
```

```python
marta.aprobado()
```
