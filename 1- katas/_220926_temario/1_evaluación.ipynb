# Asignaci√≥n y variables

La funci√≥n de la asignaci√≥n es poder reutilizar un valor varias veces y en varios sitios.

# Funciones

>La funci√≥n de una variable es la de *poder reaprovechar un valor, d√°ndole un nombre*.

>La funci√≥n de una funci√≥n (¬ølo pillas?) es *reaprovechar una **expressi√≥n**, d√°ndole un nombre*.

En el ejemplo anterior, la expresi√≥n que calcula un per√≠metro se va a repetir en varios lugares, y nosotros queremos seguir el Principio DRY a toda costa (para trabajar menos y jugar al WOW en la oficina). La soluci√≥n es utilizar una funci√≥n para darle nombre a la expresi√≥n del per√≠metro y reutilizarla siempre que queramos.

```python
def perimetro(radio):
    pi = 3.1425926
    return radio * 2 * pi
```

## El Cuerpo de la Funci√≥n

El cuerpo de la funci√≥n es todo aquel c√≥digo que se encuentra indentado bajo la declaraci√≥n, y que normalmente termina con un `return` (acordaos, ¬°s√≥lo un return en la funci√≥n, y al final del todo!).

#### √Åmbito

El cuerpo de una funci√≥n crea un **√°mbito**. Un √°mbito es un conjunto de variables. Las asignaci√≥n hechas dentro del cuerpo (como `pi` en el ejemplo de arriba) existen s√≥lo mientras se llama la funci√≥n y no son visibles desde fuera. 

Cuando se sale del cuerpo de esa funci√≥n, cualquier asignaci√≥n hecha dentro de ella (variables declaradas dentro del cuerpo de la funci√≥n) es destruida. En nuestro ejemplo, al salir de `perimetro`, `pi` deja de existir.

Desde dentro de una funci√≥n, se puede ver el **ambito** externo. Es decir, desde dentro de `per√≠metro` se puede ver cualquier otra asignaci√≥n que se haya hecho previamente.



```python
name = 'Anakin'
# desde aqu√≠ no puedo ver `pi`
def perimeter(radius):
    pi = 3.1415926
    # Desde aqu√≠, puedo ver `name`
    print(name)
    return 2 * pi * radius
    
```

#### Colisiones entre √°mbitos

Si una funci√≥n crea un nuevo √°mbito, y tambi√©n puede ver el √°mbito externo, **¬øqu√© pasa si una variable declarada en el √°mbito interno, tiene el mismo nombre que una del externo?**.

Veamos un ejemplo:

```python
pi = 45
def perimeter(radius):
    pi = 3.1415926
    # Cuanto vale pi aqui dentro?????!!!!
    return 2 * pi * radius
    
```

Cuando se produce una colisi√≥n como esa, la variable interna oculta a la externa. Es decir, en nuestro ejemplo, dentro de la funci√≥n, `pi` valdr√° `3.1415926` y el valor externo (`45`) permanecer√° oculto o inaccesible.

Cuando se salga de la funci√≥n, la asignaci√≥n anterior de `pi` volver√° a estar disponible.

> Cuando ocurre algo as√≠, se dice que la variable interna **hace sombra** a la externa.

Esto suele ocurrir por accidente y suele ser causa de bugs. Cuidado.



```python
def is_vowel(letter):
    """
    Predicado que devuelve `True` si `letter` es una vocal y `False` si no lo es.
    """
    list = ['a', 'e', 'i', 'o', 'u']
    return letter in list
```

Los par√°metros son variables que son *inyectadas* desde fuera en el √°mbito de una funci√≥n y que solo valen dentro de dicha funci√≥n.

OJO, los par√°metros tambi√©n pueden *hacer sombra* a otra variable externa.

## Reglas de Evaluaci√≥n de las Funciones

> Cuando se llama a una funci√≥n, primero se eval√∫an sus par√°metros de iquierda a derecha, y s√≥lo entonces se llama a la funci√≥n.

*Llamar a la funci√≥n* significa evaluar las expresiones del cuerpo de la funci√≥n (siguiendo las reglas ya vistas) y sustituyendo la llamada por el `valor` devuelto.

### Ejemplos



```python
def bmi(weight, height):
    """
    Peso en kilos y la altura en metros
    Devuelve en √≠ndice de masa corporal
    """
    return weight / (height * height)
```

# Expresiones Funcionales

> Una expresi√≥n funcional es una expresi√≥n que adem√°s tiene llamadas a funciones.


## Regla de Evaluaci√≥n de la Expresiones Funcionales

> Se eval√∫an siguendo las reglas de las expresiones normales, pero las llamadas a funci√≥n se eval√∫an lo primero


### Ejemplos

```python
def f(n):
    return (n * 3) 

def g(n):
    return n + 42
```
